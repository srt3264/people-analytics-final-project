```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np  
```

```python
data = pd.read_csv('/Users/suzannathompson/Desktop/People Analytics Final Paper/people-analytics-final-project/generated-data/Updated_HR_Analytics.csv')
```

```python

# Display the first few rows of the dataset
print("First few rows of the dataset:")
print(data.head())

# Display a summary of the dataset
print("\nSummary statistics of the dataset:")
print(data.describe())

# Display missing values per column
print("\nMissing values in each column:")
print(data.isnull().sum())

# Plot distributions of numerical features
num_cols = data.select_dtypes(include=['int64', 'float64']).columns
num_plots = len(num_cols)
rows = int(np.ceil(num_plots / 4))  # Adjust number of rows dynamically
plt.figure(figsize=(12, rows * 3))
for i, col in enumerate(num_cols):
    plt.subplot(rows, 4, i + 1)
    sns.histplot(data[col], kde=True)
    plt.title(f'Distribution of {col}')
plt.tight_layout()
plt.show()

# Plot count plots for categorical features
cat_cols = [col for col in data.columns if data[col].dtype == 'object' and col not in ['Question1', 'Question5']]
num_cat_plots = len(cat_cols)
rows_cat = int(np.ceil(num_cat_plots / 4))
plt.figure(figsize=(12, rows_cat * 3))
for i, col in enumerate(cat_cols):
    plt.subplot(rows_cat, 4, i + 1)
    sns.countplot(y=col, data=data)
    plt.title(f'Count plot of {col}')
    plt.tight_layout()
plt.show()

# Boxplot to check for outliers in numerical columns
plt.figure(figsize=(12, 6))
sns.boxplot(data=data.select_dtypes(include=['int64', 'float64']))
plt.title('Boxplot of numerical features')
plt.show()

# Correlation matrix of the numerical features
plt.figure(figsize=(10, 8))
sns.heatmap(data.corr(), annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
